```python
<= → subset operator

< → proper subset (means subset but not equal)

And their opposites:

>= → superset (proper superset and can be equal)

> → proper superset( canot be equal)
```

```python
proper subset (<) and proper superset (>)
a < b
Checks if all elements of a are in b AND a != b
b > a
Checks if all elements of a are in b AND a != b, 
but from the perspective of b being bigger.
```

```python
a = {1, 2}
b = {1, 2}  # can contain more element in a<b
                # 
print(a <= b)   # True (a is a proper subset if every element of A is  in B.)
print(b > a)   # True (b is a proper superset  because b contain all element of a).
```

```python
# Method 1: union() method
A = {1, 2, 3,10}
B = {3, 4, 5}
C = A.union(B)
print(C)  # Output: {1, 2, 3, 4, 5}
print(A)
```

```python
# Difference (Elements in A but not in B.)
```

```python
A = {1, 2, 3}
B = {3, 4, 5}
C = A.difference(B)
print(C)  # Output: {1, 2}

C = A - B
print(C)  # Output: {1, 2}
```

```python
# Create two sets
fruits = {"apple", "banana", "cherry"}
check_items = {"banana", "grape"}

# Empty set to store results
difference = set()

# Loop through fruits
for item in fruits:
    if item not in check_items:   # check if the item is NOT in the second set
        difference.add(item)

print("Items in fruits but not in check_items:", difference)
```

```python
fruits = {"apple", "banana", "cherry"}
check_items = {"banana", "grape"}

# Find items that are in fruits but NOT in check_items
missing = fruits - check_items  # same as fruits.difference(check_items)

print("Items in fruits but not in check_items:", missing)
```

```python
# Symmetric Difference(Elements in A or B but not both.)
```

```python
A = {1, 2, 3}
B = {3, 4, 5}

C = A.symmetric_difference(B)
print(C)  # Output: {1, 2, 4, 5}

# Method 2: ^ operator
C = A ^ B
print(C)  # Output: {1, 2, 4, 5}
```

```python
fruits = {"apple", "banana", "cherry"}
check_items = {"banana", "grape"}

# Symmetric difference using ^
sym_diff = fruits ^ check_items  # same as fruits.symmetric_difference(check_items)

print("Symmetric Difference:", sym_diff)
```

```python
fruits = {"apple", "banana", "cherry"}
check_items = {"banana", "grape"}

sym_diff = set()

# Items in fruits but not in check_items
for item in fruits:
    if item not in check_items:
        sym_diff.add(item)

# Items in check_items but not in fruits
for item in check_items:
    if item not in fruits:
        sym_diff.add(item)

print("Symmetric Difference (using loop):", sym_diff)
```

```python
copying a set creates a new set with the same elements.
```

```python
A = {1, 2, 3}
B = A.copy()  # Creates a new set with the same elements

print("previous B",B)  # Output: {1, 2, 3}

# Modifying B does NOT affect A
B.add(4)
print("after update B",B)  # Output: {1, 2, 3}
print("no effect on A",A)  # Output: {1, 2, 3, 4}
```
