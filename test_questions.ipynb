```python
#7. Given a dictionary of student marks {"Aman":85, "Riya":92, "Karan":76, "Neha":92}, find and print the name(s) and score(s) of the student(s) with the highest marks.

student_marks_dict = {
    "Aman": 85,
    "Riya" : 92,
    "Karan" : 76,
    "Neha" : 92
}

max_marks = max(student_marks_dict.values())                                                   # my solution

for i,j in student_marks_dict.items():
    if j == max_marks:
        print(i,":", j)

top_students = [name for name, score in student_marks_dict.items() if score == max_marks]         # chat gpt solution
print(max_marks, top_students)  # 92 ['Riya','Neha']
```

```python
#8.Write a program to count the frequency of each word in the sentence "python is easy and python is powerful".

from collections import Counter
sentence = "python is easy and python is powerful"
word_list = sentence.split() #will split from spaces, can also write split("")

word_count = Counter(word_list)       # Method 1 
print(word_count)

count_dict = {}                       # Method 2 
for i in word_list:
    count_dict[i] = word_list.count(i)
print(count_dict)
    
```

```python
#16.Given two dictionaries,
# d1 = {"a":10, "b":20}
# and
# d2 = {"b":5, "c":15},
# write a program to merge them. If a key is present in both, the values should be added together.

d1 = {                                                  
    "a": 10,
    "b": 20
}
d2 = {
    "b": 5,
    "c": 15
}

# Method 1 
result = d1.copy()
print(result)
for i in d2:
    if i in result:
        result[i] += d2[i]
    else:
        result[i] = d2[i]
print(result)
print(d1)

# Method 2
from collections import Counter
result = dict(Counter(d1) + Counter(d2))
print(result)
```
