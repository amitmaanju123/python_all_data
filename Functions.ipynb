```python
The isinstance() function in Python is a built-in function 
Use isinstance(obj, Class) ‚Üí check if an object belongs to a class (or subclass)
```

```python
x = 5
print(type(x))
print(isinstance(x, int))      # True
print(isinstance(x, list))    # False
```

```python
x = "amit"
print(type(x))
print(isinstance(x, (int, float)))
```

```python
def square(x):
    if not isinstance(x, (int, float)):
        raise TypeError("x must be a number")
    return x * x

print(square(5))     # ‚úÖ Works
print(square("5"))   # ‚ùå Raises TypeError
```

```python
def add(a, b):
    if isinstance(a, (int, float)) and isinstance(b, (int, float)):
        return a + b
    else:
        raise TypeError("Inputs must be numeric types")

print(add(2, 3))
print(add("1.2", 3))
```

```python
def get_length(obj):
    if isinstance(obj, (list, tuple, set)):
        return len(obj)
    elif isinstance(obj, dict):
        return len(obj.keys())
    else:
        return None
print(get_length([1, 2, 3]))  # 3
print(get_length({"a": 1}))   # 1    
```

# any

```python
Returns True if any element in an iterable is true.
Checks each element in the iterable from left to right.
Stops and returns True as soon as it finds a truthy element(true).
Returns False only if all elements are falsy.
```

```python
A value is truthy if it behaves like True

A value is falsy if it behaves like False 
# anything which is empty or 0,[],{},"",set(),false,none
```

```python
values = [0, False, 3, "", []]
print(all(values))  # ‚úÖ True (because 3 is truthy)
```

```python
values = [0, False, None, ""]
print(any(values))  # ‚ùå False (no truthy values)
```

```python
words = ["", "", "hello", ""]
print(any(words))  # True, because "hello" is a non-empty string (truthy)
```

```python
score=[1,54,87,97,56,34,98,89,90]
def get_grade(score):
    """Return a letter grade based on the numeric score."""
    if score >= 90 and score <= 100:
        return "A"
    elif score >= 80 and score < 90:
        return "B"
    elif score >= 70 and score < 80:
        return "C"
    elif score >= 60 and score < 70:
        return "D"
    elif score >= 0 and score < 60:
        return "F"
    else:
        return "Invalid score"
result=list(map(get_grade,score))   
print(result)
```

```python
def is_positive(n):
    return n > 0

nums = [-3, -1, 0, 2, 4]
positives = filter(is_positive, nums)
print(list(positives))  # [2, 4]
#üí° filter() is great for selecting data.
```

```python
numbers = [1, 2, 3, 4, 5]

# map
print(list(map(lambda x: x * 2, numbers)))    # [2, 4, 6, 8, 10]

# filter
print(list(filter(lambda x: x % 2 == 0, numbers)))  # [2, 4]
```

# Iterator( Manual control over iteration).vs Iterable

```python
An iterable in Python is any object that can return its elements one at a time,
allowing you to loop over it.
```

```python
Iterable	[1,2,3] or "abc"	Can be looped over (for x in iterable)
Lists: [1, 2, 3]

Tuples: (1, 2, 3)

Strings: "hello"

Dictionaries: {"a": 1, "b": 2}

Sets: {1, 2, 3}
```

```python
l=[1,2,3,4] 
print(type(l))
l[0]
```

```python
#When you loop over an iterable, Python internally calls iter(obj) to get an iterator, 
#then repeatedly calls next() on that iterator to get each element.

my_list = [10, 20, 30] # iterable


for item in my_list:  # my_list is iterable
    print(item)
```

```python
# You can also manually get an iterator from an iterable:
#Iterator( Manual control over iteration).
```

```python
Iterator	iter([1,2,3]) or iter("abc")	Produces elements one at a time using next()
```

```python
it = iter(my_list)  # create iterator
it
```

```python
next(it)
```

```python
# one way
print(next(it))    # 10 #initialize and store in memory
print(next(it))     # 20
print(next(it))     # 30
#returning one element at a time when next() is called.
```

```python
#2nd way
for i in it:
    print(i)
```

```python
Memory Efficiency

Iterators produce items one at a time instead of creating a full list in memory.

This is especially useful for large datasets
```

```python
Don‚Äôt store all elements at once.
Compute items only when needed.
Works with any iterable object via next()
```

# Lambda Function

```python
What is a Lambda Function in Python?

A lambda function in Python is a small, anonymous (nameless) function that is used for short, simple operations.
It is defined using the keyword lambda instead of the usual def.
```

```python
A lambda function is a quick, one-line, throwaway function ‚Äî perfect for short tasks where defining
a full function would be unnecessary.
```

```python
lambda arguments : expression
```

```python
def add(x, y):
    return x + y

add = lambda x, y: x + y
print(add(5, 3))   # Output: 8
```

```python
#. Where and Why We Use Lambda Functions

They are used when you need a simple, short function temporarily ‚Äî especially with functions like:

1.map()

2.filter()

3.reduce()

4.sorted()
```

# With map()

```python
numbers = [1, 2, 3, 4, 5]
doubled = list(map(lambda x: x * 2, numbers))
print(doubled)   # Output: [2, 4, 6, 8, 10]
```

# With filter()

```python
numbers = [1, 2, 3, 4, 5, 6]
even = list(filter(lambda x: x % 2 == 0, numbers))
print(even)   # Output: [2, 4, 6]
```

# With sorted()

```python
pairs = [(1, 3), (2, 2), (4, 1)]
sorted_pairs = sorted(pairs, key=lambda x: x[1])
print(sorted_pairs)   # Output: [(4, 1), (2, 2), (1, 3)]
```
