```python
# Q2. Write a generator function fibonacci_generator that yields Fibonacci numbers up to a given limit.

#  Using generator and next function

def fibonacci_generator(n) :
    l = [0,1]
    while ( l[-1] + l[-2] )  < n :
        a = l[-1] + l[-2] 
        l.append( a )
        yield (a)

limit = int(input("Enter the fibonacci series limit: "))
f = fibonacci_generator(limit)

for i in f:
    print(next(f))

```

```python
# Q3 Write a decorator named time_it that can be applied to any function to measure and print its execution time.
import time

def time_it(func):
    def time_counter():
        start = time.time()
        print("Execution started...")
        result = func()
        end = time.time()
        print("Execution finished in", end - start, "seconds")
        return result
    return time_counter

def execution_time():
    for i in range(1000):
        pass
        
time_it(execution_time)()           # First way 

@ time_it                           # Second way
def execution_time(): 
    for i in range(1000):
        pass

wrapped = time_it(execution_time)   # third way
wrapped()
```

```python
# Q4 Write a decorator named log_call that logs the function's name 
# and its arguments every time it is called, and then prints the result.

def log_call(func):    
    def wrapper(p,q):
        r=func(p,q)
        print(f'{r}')
    return wrapper

    

def sum1(n1,n2):
    return(f"sum1,{n1},{n2}")

wrapped = log_call(sum1)
wrapped(3,4)
       
```

```python
# Q10. Write a Python script using the re module to extract all 
# valid email addresses from the string "My emails are test@mail.com and hello123@gmail.com".
import re

email_addr = "My emails are test@mail.com and hello123@gmail.com"

# emails = re.findall(r'[A-Za-z0-9._%+-]+@[A-Za-z0-9.-]+\.[A-Za-z]{2,}', email_addr)
emails = re.findall(r'\S+@\S+', email_addr)
# [A-Za-z0-9._%+-]+       → username part (letters, numbers, ., _, %, +, -)
# @                       → @ symbol
# [A-Za-z0-9.-]+          → domain name
# \.                      → dot
# [A-Za-z]{2,}            → extension (com, in, org, etc.)

print("Valid Emails are: ",emails)
```

```python
# Q12.Write a function is_valid_password(password) that uses regex to validate 
# if a password meets these criteria: at least 1 uppercase letter, 1 lowercase letter, 1 digit, 
# and a minimum length of 8 characters. The function should return True or False.
import re
def is_valid_password(password):
    pattern = r'^(?=.*[A-Z])(?=.*[a-z])(?=.*\d).{8,}$'
    
    # ?                 → Zero or one time (making it optional).
    # ^(?=.*[A-Z])      → at least one uppercase
    # (?=.*[a-z])       → at least one lowercase
    # (?=.*\d)          → at least one digit
    # .{8,}$            → minimum length 8

    match = re.match(pattern, password)
    return bool(match)

password = input("Enter your password: ")

if is_valid_password(password): 
    print(True)
else : 
    print(False)
    
```
